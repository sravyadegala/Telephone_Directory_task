# **Telephone Directory :**
- Create a simple user define function which will check for the mobile number for a given name.
- If the name does not exist, it should ask for the adding, into the database.
- We should give multiple options like:
- Display numbers, Add phone number, Search number and Quit
- If the user select quit then only program should terminate, otherwise the program should run.


```python
class TelephoneDirectory:
    def __init__(self):
        self.contacts={}
        
        
    def display_number(self):
        print("***********Telephone Numbers***************")
        for name,number in self.contacts.items():
            print(f"Name:{name} Number:{number}")
        
        
    def add_number(self):
        name=input("Enter your name:  ")
        number=input("Enter your number:91 ")
        self.contacts[name]=number
        print("Sucessfully added your contact")

    
    def search_number(self):
        name=input("Enter Name you are searching for:  ")
        if name in self.contacts:
            print(self.contacts[name])
        else:
            print(f"No Name found for {name}")
            choice=input("Should add to contacts?  (y/n):")
            if choice=='y':
                self.add_number()

    def Directory(self):
        while True:
            print("1. Display phone number")
            print("2. Add a phone number")
            print("3. Search contact")
            print("Quit")
            
            select=input("type in numbers (1-4) : ") 
            if select=='1':
                self.display_number()
            elif select=='2':
                self.add_number()
            elif select=='3':
                self.search_number()
            elif select=='4':
                break

            else:
                print("Invalid choice,please try again")

d1=TelephoneDirectory()
d1.Directory()
```
